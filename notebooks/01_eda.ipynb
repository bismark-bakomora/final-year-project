```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('../data/raw/heart_statlog_cleveland_hungary_final.csv')
```

```python
FEATURE_COLS = [
    'age', 'sex', 'chest pain type', 'resting bp s',
    'cholesterol', 'fasting blood sugar', 'resting ecg',
    'max heart rate', 'exercise angina', 'oldpeak', 'ST slope'
]

# Correlation heatmap for features and target
plt.figure(figsize=(14, 10))
corr = df[FEATURE_COLS + ['target']].corr()
sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.5
)
plt.title('Pairwise Correlation Heatmap')
plt.tight_layout()
plt.savefig('../outputs/figures/correlation_heatmap.png', dpi=150)
plt.show()

# Print key correlations with target
print("\nCorrelation with target:")
print(corr['target'].sort_values(ascending=False).round(2))
```
